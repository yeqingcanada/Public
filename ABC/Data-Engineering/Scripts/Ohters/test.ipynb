```python
import pandas as pd


data = pd.read_csv(r'calculation_input_df.csv')
df = pd.DataFrame(data)

result_df = df.groupby(['ProductFamilyL1', 'MaterialType', 'MaterialStdUOM', 'CapacityAreaID', 'Plant', 'FormAssembly'], dropna=False).agg(
    InvQty=('InvQty', 'sum'),
    InvPallets=('InvPallets', 'sum'),
    YearThroughput=('YearThroughput', 'sum'),
    DayThroughput=('DayThroughput', 'sum')
).reset_index()

result_df
```
